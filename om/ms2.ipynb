```python

class Price:
  """Price class."""
  def __init__(self, part_number: int, price: float):
    self.price = price
    self.part_number = part_number

  def get_price(self):
    return f"{self.price:.2f}"


def set_discount(item_price, percent_off: float):
  """Adds a 'percent_off' attribute to the item_price"""
  item_price.percent_off = percent_off


def get_discount_price(item_price):
  """Calculates the discount price."""
  return f"{item_price.price - (item_price.price * (item_price.percent_off/100)):.2f}"


item_price = Price(8893, 13.17)
```

```python
print(f"dir(Price):\n{dir(Price)}\n\n")
print(f"dir(item_price):\n{dir(item_price)}")
```

```python
print(f"Price.__dict__:\n{Price.__dict__}\n\n")
print(f"item_price.__dict__:\n{item_price.__dict__}")
```

```python
Price.set_discount = set_discount
print(f"dir(Price):\n{dir(Price)}\n")

item_price.set_discount(15)

Price.get_discount_price = get_discount_price
print(f"dir(Price):\n{dir(Price)}\n\n")
print(f"item_price.get_discount_price:\n{item_price.get_discount_price}\n")

print(f"the discounted price is: {item_price.get_discount_price()}")
```
