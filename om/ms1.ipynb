```python
class Price:
  """Price class."""
  def __init__(self, part_number, price):
    self.price = price
    self.part_number = part_number

  def get_price(self):
    return f"{self.price:.2f}"


def function__init__(self, part_number, price):
  self.price = price
  self.part_number = part_number


def function_get_price(self):
  return f"{self.price:.2f}"


p = Price(324521, 69.0023459876)
print(f"pn: {p.part_number} - ${p.price:.2f}")
print(p.get_price())

namespace = {
  "__init__": function__init__,
  "get_price": function_get_price,
}

Price2 = type("Price2", (), namespace)
p2 = Price2(324521, 68.9998795)
print(f"pn: {p2.part_number} - ${p2.price:.2f}")
print(p2.get_price())
```
