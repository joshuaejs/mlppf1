```python
my_int: int = 42
another_int: int = my_int
print(f"my_int is at address: {id(my_int)} and has the value: {my_int}")
print(f"another_int is at address: {id(another_int)} and has the value: {another_int}")
# this will change the memory address
another_int = 254
print()
print(f"my_int is at address: {id(my_int)} and has the value: {my_int}")
print(f"another_int is now at address: {id(another_int)} and has the value: {another_int}")
```

```python
my_float: float = 0.42
another_float: float = my_float
print(f"my_float is at address: {id(my_float)} and has the value: {my_float}")
print(f"another_float is at address: {id(another_float)} and has the value: {another_float}")
# this will change the memory address
another_float = 3.141519
print()
print(f"my_float is at address: {id(my_float)} and has the value: {my_float}")
print(f"another_float is now at address: {id(another_float)} and has the value: {another_float}")
```

```python
my_str: str = "a string"
another_str: str = my_str
print(f"my_str is at address: {id(my_str)} and has the value: {my_str}")
print(f"another_str is at address: {id(another_str)} and has the value: {another_str}")
# strings do not support item assignment
# this will change the memory address
another_str = "a string"
print()
print(f"my_str is at address: {id(my_str)} and has the value: {my_str}")
print(f"another_str is now at address: {id(another_str)} and has the value: {another_str}")
```

```python
my_list: list = [ 13, "t", 24 ]
another_list: list = my_list
print(f"my_list is at address: {id(my_list)} and has the value: {my_list}")
print(f"another_list is at address: {id(another_list)} and has the value: {another_list}")
print()
# this will not change the memory address, only the element
another_list[2] = 200
print(f"my_list is at address: {id(my_list)} and has the value: {my_list}")
print(f"another_list is at address: {id(another_list)} and has the value: {another_list}")
print()
# this creates a new list
my_list = [ "a", "new", "list" ]
print(f"my_list is at address: {id(my_list)} and has the value: {my_list}")
print(f"another_list is at address: {id(another_list)} and has the value: {another_list}")
```

```python
my_tuple: tuple = (13, "t", 24)
another_tuple: tuple = my_tuple
print(f"my_tuple is at address: {id(my_tuple)} and has the value: {my_tuple}")
print(f"another_tuple is at address: {id(another_tuple)} and has the value: {another_tuple}")
# tuples do not support item assignment
# this will change the memory address
another_tuple = ("spam", "strawberries", 3.141519)
print()
print(f"my_tuple is at address: {id(my_tuple)} and has the value: {my_tuple}")
print(f"another_tuple is now at address: {id(another_tuple)} and has the value: {another_tuple}")
```

```python
my_dict: dict = {
  "a_key": "a value",
  13: 42,
}
another_dict: dict = my_dict
print(f"my_dict is at address: {id(my_dict)} and has the value: {my_dict}")
print(f"another_dict is at address: {id(another_dict)} and has the value: {another_dict}")
# this will only update the element
another_dict["a_key"] = 63543
print()
print(f"my_dict is at address: {id(my_dict)} and has the value: {my_dict}")
print(f"another_dict is at address: {id(another_dict)} and has the value: {another_dict}")
```
